```python
# =============================================================================
# SCRIPT 1: Custom CNN Baseline for Fungal Keratitis vs. Normal Cornea
# =============================================================================

FUNGAL_PATH     = "/content/drive/MyDrive/FK/FK pic"
NORMAL_PATH     = "/content/drive/MyDrive/normal/normal"
MODEL_SAVE_PATH = "/content/drive/MyDrive/best_custom_cnn.keras"

import os
import shutil
import tempfile
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_curve, auc,
    precision_recall_curve, average_precision_score,
)

SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)

IMG_SIZE   = (224, 224)
BATCH_SIZE = 32
EPOCHS     = 10
LR         = 1e-3
VAL_SPLIT  = 0.20
DROPOUT    = 0.5

TF_SUPPORTED_FORMATS = {".jpg", ".jpeg", ".png", ".bmp", ".gif", ".webp"}

# =============================================================================
# 1.  PRE-FLIGHT — validate files with TF decoder (not just PIL)
# =============================================================================
# KEY FIX: PIL.verify() only checks the file header, not the full decode.
# Files can have a supported extension but an incompatible internal format
# (e.g. a TIFF saved as .jpg). We now attempt a real TF decode to be sure.

def is_tf_decodable(fpath: str) -> bool:
    """Return True only if TF can fully decode the image."""
    try:
        raw = tf.io.read_file(fpath)
        img = tf.io.decode_image(raw, channels=3, expand_animations=False)
        _ = img.numpy()          # force eager execution / actual decode
        return True
    except Exception:
        return False

print("\n[PRE-FLIGHT]  Scanning and validating source folders …")
print(f"   Scanning: {FUNGAL_PATH}")
print(f"   Scanning: {NORMAL_PATH}\n")

all_files     = {"fungal": [], "normal": []}
skipped_files = []

for label, folder in [("fungal", FUNGAL_PATH), ("normal", NORMAL_PATH)]:
    for root, _, files in os.walk(folder):
        for fname in files:
            if fname.startswith("."):
                skipped_files.append((os.path.join(root, fname), "hidden file"))
                continue
            ext   = os.path.splitext(fname)[1].lower()
            fpath = os.path.join(root, fname)
            if ext not in TF_SUPPORTED_FORMATS:
                skipped_files.append((fpath, f"unsupported extension: '{ext}'"))
                continue
            # ── THE FIX: validate with TF, not PIL ──
            if is_tf_decodable(fpath):
                all_files[label].append(fpath)
            else:
                skipped_files.append((fpath, "TF cannot decode (bad format/corrupt)"))

print(f"   Valid files  →  fungal: {len(all_files['fungal'])}  |  normal: {len(all_files['normal'])}")
if skipped_files:
    print(f"\n⚠️  Skipping {len(skipped_files)} incompatible/corrupt files:")
    for fpath, reason in skipped_files[:20]:
        print(f"   [{reason}]  {os.path.basename(fpath)}")
    if len(skipped_files) > 20:
        print(f"   … and {len(skipped_files) - 20} more")
else:
    print("   ✅  No incompatible files found")

# ── Build clean temp directory ────────────────────────────────────────────────
print("\n   Building clean temp directory …")

temp_dir    = tempfile.mkdtemp(prefix="keratitis_clean_")
fungal_dest = os.path.join(temp_dir, "fungal")
normal_dest = os.path.join(temp_dir, "normal")
os.makedirs(fungal_dest, exist_ok=True)
os.makedirs(normal_dest, exist_ok=True)

def copy_files(src_list, dest_dir):
    for src in src_list:
        dst = os.path.join(dest_dir, os.path.basename(src))
        if os.path.exists(dst):
            base, ext = os.path.splitext(os.path.basename(src))
            dst = os.path.join(dest_dir, f"{base}_{abs(hash(src))}{ext}")
        shutil.copy2(src, dst)

copy_files(all_files["fungal"], fungal_dest)
copy_files(all_files["normal"], normal_dest)

n_fungal = len(all_files["fungal"])
n_normal = len(all_files["normal"])

print(f"   ✅  Clean temp root : {temp_dir}")
print(f"   fungal/ ← {n_fungal} files")
print(f"   normal/ ← {n_normal} files")

# =============================================================================
# 2.  DATA LOADING
# =============================================================================
print("\n[1/6]  Loading datasets …")

train_ds = tf.keras.utils.image_dataset_from_directory(
    temp_dir,
    validation_split=VAL_SPLIT,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="binary",
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    temp_dir,
    validation_split=VAL_SPLIT,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="binary",
)

CLASS_NAMES = train_ds.class_names
print(f"   Classes : {CLASS_NAMES}")

def ensure_rgb_batch(images, labels):
    channels = tf.shape(images)[-1]
    images = tf.cond(
        tf.equal(channels, 1),
        true_fn=lambda: tf.repeat(images, repeats=3, axis=-1),
        false_fn=lambda: tf.cond(
            tf.equal(channels, 4),
            true_fn=lambda: images[:, :, :, :3],
            false_fn=lambda: images,
        ),
    )
    images = tf.ensure_shape(images, (None, IMG_SIZE[0], IMG_SIZE[1], 3))
    return images, labels

AUTOTUNE = tf.data.AUTOTUNE

train_ds = (
    train_ds
    .map(ensure_rgb_batch, num_parallel_calls=AUTOTUNE)
    .cache()
    .shuffle(1000, seed=SEED)
    .prefetch(AUTOTUNE)
)

val_ds = (
    val_ds
    .map(ensure_rgb_batch, num_parallel_calls=AUTOTUNE)
    .cache()
    .prefetch(AUTOTUNE)
)

print("\n   🔍  Batch sanity check …")
for images, labels in train_ds.take(1):
    print(f"   Batch image shape : {images.shape}")
    print(f"   Batch label shape : {labels.shape}")
    print(f"   Pixel range       : [{images.numpy().min():.1f}, {images.numpy().max():.1f}]")
    print(f"   Unique labels     : {np.unique(labels.numpy())}")
print("   ✅  Batch check passed\n")

# =============================================================================
# 3.  CLASS WEIGHTS
# =============================================================================
print("\n[2/6]  Computing class weights …")

total_samples = n_fungal + n_normal
counts        = [n_fungal, n_normal]
class_weights = {i: total_samples / (2 * counts[i]) for i in range(2)}
print(f"   Counts  : { {CLASS_NAMES[i]: counts[i] for i in range(2)} }")
print(f"   Weights : { {CLASS_NAMES[i]: round(class_weights[i], 4) for i in range(2)} }")

# =============================================================================
# 4.  AUGMENTATION
# =============================================================================
augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical", seed=SEED),
    tf.keras.layers.RandomRotation(0.15, seed=SEED),
    tf.keras.layers.RandomZoom(0.15, seed=SEED),
], name="augmentation")

# =============================================================================
# 5.  MODEL — 5-Block Custom CNN
# =============================================================================
print("\n[3/6]  Building Custom CNN …")

def conv_block(x, filters, kernel_size=3):
    x = tf.keras.layers.Conv2D(filters, kernel_size, padding="same", use_bias=False)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)
    x = tf.keras.layers.MaxPooling2D(pool_size=2)(x)
    return x

inputs = tf.keras.Input(shape=(*IMG_SIZE, 3), name="input_image")
x      = augmentation(inputs)
x      = tf.keras.layers.Rescaling(1.0 / 255)(x)

x = conv_block(x, 32)
x = conv_block(x, 64)
x = conv_block(x, 128)
x = conv_block(x, 256)
x = conv_block(x, 512)

x       = tf.keras.layers.GlobalAveragePooling2D()(x)
x       = tf.keras.layers.Dropout(DROPOUT)(x)
outputs = tf.keras.layers.Dense(1, activation="sigmoid", name="prediction")(x)

model = tf.keras.Model(inputs, outputs, name="Custom_CNN_Baseline")
model.summary()

# =============================================================================
# 6.  COMPILE & CALLBACKS
# =============================================================================
print("\n[4/6]  Compiling …")

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LR),
    loss="binary_crossentropy",
    metrics=[
        "accuracy",
        tf.keras.metrics.AUC(name="auc"),
        tf.keras.metrics.Precision(name="precision"),
        tf.keras.metrics.Recall(name="recall"),
    ],
)

callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor="val_loss", patience=10,
        restore_best_weights=True, verbose=1,
    ),
    tf.keras.callbacks.ModelCheckpoint(
        filepath=MODEL_SAVE_PATH, monitor="val_loss",
        save_best_only=True, verbose=1,
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss", factor=0.5,
        patience=5, min_lr=1e-7, verbose=1,
    ),
]

# =============================================================================
# 7.  TRAINING
# =============================================================================
print("\n[5/6]  Training …")

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1,
)

# =============================================================================
# 8.  EVALUATION & PLOTS
# =============================================================================
print("\n[6/6]  Evaluating …")

y_true, y_prob = [], []
for images, labels in val_ds:
    probs = model(images, training=False).numpy().flatten()
    y_prob.extend(probs)
    y_true.extend(labels.numpy().flatten())

y_true = np.array(y_true)
y_prob = np.array(y_prob)
y_pred = (y_prob >= 0.5).astype(int)

print("\n── Classification Report ────────────────────────────────────────────")
print(classification_report(y_true, y_pred, target_names=CLASS_NAMES, digits=4))

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle(
    "Custom CNN Baseline — Evaluation Dashboard\n"
    f"fungal: {n_fungal}  |  normal: {n_normal}  |  val split: {int(VAL_SPLIT*100)}%",
    fontsize=13, fontweight="bold",
)

# Accuracy
ax = axes[0, 0]
ax.plot(history.history["accuracy"],     label="Train", lw=2, color="steelblue")
ax.plot(history.history["val_accuracy"], label="Val",   lw=2, color="steelblue", ls="--")
ax.set_title("Accuracy", fontweight="bold"); ax.set_xlabel("Epoch"); ax.set_ylabel("Accuracy")
ax.legend(); ax.grid(alpha=0.3)

# Loss
ax = axes[0, 1]
ax.plot(history.history["loss"],     label="Train", lw=2, color="darkorange")
ax.plot(history.history["val_loss"], label="Val",   lw=2, color="darkorange", ls="--")
ax.set_title("Loss", fontweight="bold"); ax.set_xlabel("Epoch"); ax.set_ylabel("BCE Loss")
ax.legend(); ax.grid(alpha=0.3)

# ROC
ax = axes[0, 2]
fpr, tpr, _ = roc_curve(y_true, y_prob)
auroc = auc(fpr, tpr)
ax.plot(fpr, tpr, color="steelblue", lw=2, label=f"AUROC = {auroc:.4f}")
ax.plot([0, 1], [0, 1], "k--", lw=1)
ax.fill_between(fpr, tpr, alpha=0.08, color="steelblue")
ax.set_title("ROC Curve", fontweight="bold"); ax.set_xlabel("FPR"); ax.set_ylabel("TPR")
ax.legend(loc="lower right"); ax.grid(alpha=0.3)

# PR Curve
ax = axes[1, 0]
prec_vals, rec_vals, _ = precision_recall_curve(y_true, y_prob)
auprc = average_precision_score(y_true, y_prob)
ax.plot(rec_vals, prec_vals, color="darkorange", lw=2, label=f"AUPRC = {auprc:.4f}")
ax.fill_between(rec_vals, prec_vals, alpha=0.08, color="darkorange")
ax.set_title("Precision-Recall Curve", fontweight="bold")
ax.set_xlabel("Recall"); ax.set_ylabel("Precision")
ax.legend(); ax.grid(alpha=0.3)

# Confusion Matrix
ax = axes[1, 1]
cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES,
            linewidths=0.8, annot_kws={"size": 16, "weight": "bold"}, ax=ax)
ax.set_title("Confusion Matrix", fontweight="bold")
ax.set_xlabel("Predicted"); ax.set_ylabel("True")

# Per-Class Metrics
ax = axes[1, 2]
report      = classification_report(y_true, y_pred, target_names=CLASS_NAMES, output_dict=True)
groups      = CLASS_NAMES + ["macro avg"]
metric_keys = ["precision", "recall", "f1-score"]
bar_colors  = ["steelblue", "darkorange", "seagreen"]
bar_data    = np.array([[report[g][m] for m in metric_keys] for g in groups])
x_pos       = np.arange(len(groups))
bar_width   = 0.25

for idx, (metric, color) in enumerate(zip(metric_keys, bar_colors)):
    bars = ax.bar(x_pos + idx * bar_width, bar_data[:, idx],
                  bar_width, label=metric.capitalize(), color=color, alpha=0.85)
    for bar in bars:
        h = bar.get_height()
        ax.annotate(f"{h:.2f}",
                    xy=(bar.get_x() + bar.get_width() / 2, h),
                    xytext=(0, 3), textcoords="offset points",
                    ha="center", va="bottom", fontsize=7)

ax.set_xticks(x_pos + bar_width)
ax.set_xticklabels(groups, rotation=15, fontsize=9)
ax.set_ylim(0, 1.15); ax.set_ylabel("Score")
ax.set_title("Per-Class Metrics", fontweight="bold")
ax.legend(); ax.grid(axis="y", alpha=0.3)

plt.tight_layout()
plot_save_path = "/content/drive/MyDrive/custom_cnn_evaluation.png"
plt.savefig(plot_save_path, dpi=150, bbox_inches="tight")
plt.show()

print("\n" + "=" * 60)
print("  TRAINING COMPLETE")
print("=" * 60)
print(f"  AUROC          : {auroc:.4f}")
print(f"  AUPRC          : {auprc:.4f}")
print(f"  Epochs trained : {len(history.history['loss'])}")
print(f"  Best val loss  : {min(history.history['val_loss']):.4f}")
print(f"  Best val acc   : {max(history.history['val_accuracy']):.4f}")
print(f"  Model saved    : {MODEL_SAVE_PATH}")
print(f"  Plot  saved    : {plot_save_path}")
print("=" * 60)
```
